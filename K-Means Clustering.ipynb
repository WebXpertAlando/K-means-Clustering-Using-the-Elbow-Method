# What is K-Means Clustering Algorithm
<hr>
K-Means clustering is an unsupervised learning algorith, which groups the unlabeled dataset into different clusters. Here **k** defines the number of pre-defined clusters that need to be created in the line process, as if K=2, there will be two clusters, and K=3 it means that there will be clusters, and so on.

>It is an iterative algorithm that divides the unlabeled dataset into **k** different clusters in such a way that each dataset belongs only to only one group that has similar properties

It allows us to cluster the data into different groups and a convinient way to discover the catogories of groups in the unlabeled dataset on its own without the need for any training.

It is a centroid-based algorithm, where clusters is associated with centroid. The main aim of this algorithm is to minimize the sum  of distances between the data point and the their correspoing clusters. 

The algorithm takes the unlabeled dataset as input, divides the dataset into **k-number** of cluster, and repeats the process until it does not find the best clusters. The value of **k** should be predetermined in this algorithm.

The k-means clustering algorithm mainly performs two tasks:
  - Determines the best value for **k** center points or centroids by an iterative process.
  - Assigns each data point to its closest k-center. Those data points which are near to the particular k-center, create a cluster.


Hence each cluster has data points with some commonalities, and it is away from other clusters. Consider the below diagram that explains the working of the kK-means Clustering Algorithm.






# How does the K-Means Algorithm Work?

 - Select the number of **k** to decide the number of cluster.
 
 - Select random K points or centroids (it can be either from the input data set).
   
 - Assign each data point to their closet centroid, which will form the predefined K clusters.
   
 - Calculate the variance and place a new centroid of each cluster.
   
 - Repeat the third step which means, reassign each data point to the new closet centroid of each cluster.
   
 - If any reassignment occurs, then go to step 4 else go to finish.

# How to find choose the value of "k" number of clusters in Kmeans Clustering
The performance of the K-means clustering algorithm depends upon highly efficinet clusters that it forms. But choosing the optimal number of clusters is a big task. 

There are some different ways to find the optimaal numbers of clusters, but here we are discussing the most appropriate method  to find the number of clusters or values of **k**. The example method is given below:

The **Elbow** is one of the most popular ways to find the optimal number of clusters. This method uses the concept of **Inertia** value. To measure the distance of between data points and centroids, we can use any method such as **Euclidean** distance or **Manhattan** distance.

To find the optimal value of clusters, the elbow method follows the below steps:

- It executes the K-means clustering on a given dataset for different **k** values(ranges from 1-10).

- For each value in our **k**, it calculates the **inertia_** value.

- It plot a curve between calculated inertia_ values and the number of **k** clusters.

- The sharp point of bend or a point of the plot looks like an arm, them that plot is considered as the best value of **k**.

Since the graph shaows the sharp bend, which looks like an elbow, hene it is known as the **Elbow** method.

# Standardization and Feature Scaling

```python
scaler = StandardScaler()

X = df_customers.iloc[:,[3,4]]
```

# Fitting the Training and Transformation

```python
X_scaled = scaler.transform(X)
```

```python
X_scaled.mean(axis=0)
```

# Training the Kmeans Algorithm on the Training Dataset
To train the model, we will use 5 as know there are five clusters thst need to be formed. 

```python
# Instantiate the model
kmeans = KMeans(n_clusters=5, random_state=42)
# fit the model
kmeans.fit(X_scaled)
```

# Make Prediction 

```python
# predict
y_pred = kmeans.fit_predict(X_scaled)

print(y_pred)
```

By executing the above lines of code, we are using the **y_pred** variable. We can now check the values of **y_pred** with our original dataset. From the above results, these prediction list each value of prediction belongs to a cluster. We can say that a customer who has a customer_id 1 belongs to cluster 2 and so on.

# Inspect the coordinates of the 5 centroids

```python
kmeans.cluster_centers_
```

As we have got the number of cluster and centroids, so we can now find the optimal k number of clusters by using the **Elbow Method**

# Finding the Optimal (k) number of clusters using the Elbow Method 

```python
kmeans  = KMeans(n_clusters=3)
kmeans.fit(X_scaled)
kmeans.inertia_
```

Or when k = 5, 

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(X_scaled)
kmeans.inertia_
```

```python
# Plot the k of interia
ks = range(1,11)
inertias = []

for k in ks:
    model=KMeans(n_clusters=k)
    model.fit(X_scaled)
    inertias.append(model.inertia_)

plt.plot(ks,inertias, '-o')
plt.xlabel('Number of Clusters, k')
plt.ylabel('Inertia')
plt.xticks(ks)
plt.show()
```

From the above pot, we can see that the elbow point is that **5** so the number of clusters here will be **5**

# Visualizing the Clusters
Next, we will visualize the clusters. As we have 5 clusters for our model, so we will visualize each cluster one by one. To visualize the clusters, we will use scatter plot.

```python
# Plot the scaled data
plt.scatter(X_scaled[:,0],
            X_scaled[:,1],
            c=y_pred)
# Identify the centroids
plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1],
            marker="o",
            s=250,
            c=[0,1,2,3,4],
            edgecolors='k')
plt.xlabel('Annual Income'); plt.ylabel('Spending Score')
plt.title('Kmeans (k=5)')
plt.show()
```
